```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4],[10,20,30,40])
plt.show()
```

```python
plt.plot([1,2,3,4],[10,20,30,40])
plt.title("Data visulaization")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot([1,2,3,4],[10,20,30,40])
plt.show()
```

```python
plt.plot([1,2,3,4],[10,20,30,40],"go")
plt.title("Data visulaization")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x = np.arange(1,5)
y = x**5
plt.plot([1,2,3,4],[10,20,30,40],"go",x,y,"r^")
plt.title("Data visulaization")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x = np.arange(0,6,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title("Data visulaization")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
months = ["jan","feb","mar","apr","may","jun"]
sales = [100,200,150,300,450,70]
plt.bar(months,sales,color = "red")
plt.title("Data visulaization Bar Graph")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
months = ["jan","feb","mar","apr","may","jun"]
sales = [100,200,150,300,450,70]
plt.barh(months,sales,color = "red")
plt.title("Data visulaization Horizontally Bar Graph")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```
